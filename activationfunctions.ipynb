```python
import numpy as np
import matplotlib.pyplot as plt

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # subtracting max(x) for numerical stability
    return e_x / e_x.sum(axis=0)

# Generate some sample data for demonstration (one class)
x = np.linspace(-5, 5, 100)
scores = x  # Scores for one class

# Apply softmax to the scores
softmax_scores = softmax(scores)

# Plotting
plt.figure(figsize=(8, 6))

# Plot softmax probabilities
plt.plot(x, softmax_scores, label='Softmax Probabilities')

plt.title('Softmax Function for One Class')
plt.xlabel('Input Value (Logits)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Rectified Linear Unit (ReLU)
def ReLU(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)

# Generating data for Graph
x_data = np.linspace(-10,10,100)
y_data = ReLU(x_data)
# dy_data = der_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('ReLU Activation Function & Derivative')
plt.legend(['ReLU'])
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)

# Generating data For Graph
x_data = np.linspace(-10,10,100)
y_data = leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('leaky ReLU Activation Function')
plt.legend(['leaky_ReLU'])
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbolic Tangent (htan) Activation Function
def htan(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
    # htan derivative
def der_htan(x):
    return 1 - htan(x) * htan(x)

# Generating data for Graph
x_data = np.linspace(-6,6,100)
y_data = htan(x_data)
dy_data = der_htan(x_data)
plt.plot(x_data, dy_data)
plt.title('htan Activation Function & Derivative')
plt.legend(['der_htan'])
plt.grid()
plt.show()
```
